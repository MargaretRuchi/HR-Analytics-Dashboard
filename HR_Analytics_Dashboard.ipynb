```python
print("Total Employees:", len(df))
```

```python
print("Employees Left:", df[df['Attrition']=='Yes'].shape[0])
```

```python
attrition_rate = (
    df[df['Attrition']=='Yes'].shape[0]
    / len(df)
) * 100

print("Attrition Rate:", round(attrition_rate,2), "%")
```

```python
print(
    "Average Monthly Income:",
    round(df['MonthlyIncome'].mean(),2)
)
```

```python
print(
    "Average Employee Age:",
    round(df['Age'].mean(),2)
)
```

```python
plt.figure(figsize=(8,5))

sns.countplot(
    data=df,
    x='Attrition'
)

plt.title('Employee Attrition Distribution')

plt.savefig(
    'attrition_distribution.png',
    bbox_inches='tight'
)

plt.show()
```

```python
department_attrition = pd.crosstab(
    df['Department'],
    df['Attrition']
)

department_attrition
```

```python
department_attrition.plot(
    kind='bar',
    figsize=(10,6)
)

plt.title('Department-wise Attrition')

plt.ylabel('Employee Count')

plt.savefig(
    'department_attrition.png',
    bbox_inches='tight'
)

plt.show()
```

```python
plt.figure(figsize=(10,6))

sns.histplot(
    df['MonthlyIncome'],
    bins=30
)

plt.title('Monthly Income Distribution')

plt.savefig(
    'income_distribution.png',
    bbox_inches='tight'
)

plt.show()
```

```python
plt.figure(figsize=(10,6))

sns.boxplot(
    data=df,
    x='Attrition',
    y='Age'
)

plt.title('Age Distribution by Attrition')

plt.savefig(
    'age_attrition.png',
    bbox_inches='tight'
)

plt.show()
```

```python
job_satisfaction = pd.crosstab(
    df['JobSatisfaction'],
    df['Attrition']
)

job_satisfaction
```

```python
job_satisfaction.plot(
    kind='bar',
    figsize=(10,6)
)

plt.title('Job Satisfaction vs Attrition')

plt.ylabel('Employee Count')

plt.savefig(
    'job_satisfaction_attrition.png',
    bbox_inches='tight'
)

plt.show()
```

```python
overtime_attrition = pd.crosstab(
    df['OverTime'],
    df['Attrition']
)

overtime_attrition
```

```python
overtime_attrition.plot(
    kind='bar',
    figsize=(8,5)
)

plt.title('Overtime vs Attrition')

plt.ylabel('Employee Count')

plt.savefig(
    'overtime_attrition.png',
    bbox_inches='tight'
)

plt.show()
```

```python
summary = pd.DataFrame({
    'Metric': [
        'Total Employees',
        'Employees Left',
        'Attrition Rate'
    ],
    'Value': [
        len(df),
        df[df['Attrition']=='Yes'].shape[0],
        round(attrition_rate,2)
    ]
})

summary
```
